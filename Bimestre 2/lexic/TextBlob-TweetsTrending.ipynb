## Carga de datos del DF

## Detecci√≥n del idioma e inserci√≥n del resultado en una nueva columna del df

## Advertencia sobre el l√≠mite de uso de Google Cloud
> No se deben realizar demasiadas peticiones al mismo tiempo, se recomienda establecer una espera entre cada petici√≥n para evitar el HTTP Error 429: Too Many Requests.

```python
import time

# Creamos nuevas columnas
df = df.assign( idioma = "" )
df = df.assign( polaridad = "" )
df = df.assign( subjetividad = "" )

index = 0
print('Tama√±o del df: ' , df.shape)

for texto in df['texto']:
    # Impresi√≥n del √≠ndice que estamos tratando
    print(index , '\n')
    
    tweet = TextBlob(texto)
    
    # Detecci√≥n del idioma
    idiomaDetectado = tweet.detect_language()
    
    # Detecci√≥n de la polaridad
    polaridad = tweet.sentiment.polarity
    
    # Detecci√≥n de la subjetividad
    subjetividad = tweet.sentiment.subjectivity
    
    print('Language: ' , idiomaDetectado , ' Polaridad: ' , polaridad , ' Subjetividad: ' , subjetividad , ': \n' , texto)
    
    # Guardamos los resultados
    df.loc[index, 'idioma'] = idiomaDetectado
    df.loc[index, 'polaridad'] = polaridad
    df.loc[index, 'subjetividad'] = subjetividad
    
    # Pasamos a la siguiente fila
    index = index + 1
    
    time.sleep(1)
```

```python
df.groupby(['idioma']).count()
```

```python
df.groupby(['polaridad']).count()
```

```python
df.groupby(['subjetividad']).count()
```
