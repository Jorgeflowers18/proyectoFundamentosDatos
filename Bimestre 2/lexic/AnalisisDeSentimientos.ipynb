## An√°lisis del idioma, polaridad y subjetividad

# Carga del dataset

```python
import pandas

df = pandas.read_csv('TweetsTrendingEc-8Julio21conNLP.csv', sep = ';')
df
```

### Eliminamos aquellos registros que carecen de polaridad o subjetividad que analizar 

```python
# Creamos la condici√≥n
df = df[df.polaridad != 0]
df = df[df.subjetividad != 0]
df
```

```python
df.groupby('topic').nunique()
```

### Comparamos aquellos tweets con mayor y menor polaridad

La polaridad es un valor flotante que se encuentra en el rango [-1.0, 1.0]

```python
porPolaridadInversa = df.sort_values('polaridad', ascending = False)    # Textos mas cercanos a una opini√≥n positiva
porPolaridadInversa.head(20)
```

```python
porPolaridad = df.sort_values('polaridad', ascending = True)    # Textos mas cercanos a una opini√≥n negativa
porPolaridad.head(20)
```

### Tweets con mayor subjetividad
La subjetividad es un valor flotante comprendido en el rango [0.0, 1.0] donde 0.0 es muy objetivo (definici√≥n) y 1.0 es muy subjetivo (opinici√≥n personal).

```python
porSubjetividad = df.sort_values('subjetividad', ascending = False) # Textos m√°s cercanos a una definici√≥n
porSubjetividad.head(20)
```

```python
porSubjetividadInversa = df.sort_values('subjetividad', ascending = True)   # Textos mas cercanos a una opini√≥n personal
porSubjetividadInversa.head(20)
```

## Gr√°ficas
La columna subjetividad est√° comprendida en el rango [0 to 1], donde 1 representa a un tweet m√°s cercano a una opini√≥n personal mientras que su valor contrario hace referencia a un texto mas objetivo, cercano a una definici√≥n.

La columna polaridad est√° comprendida en el rango [1 to -1], donde los valores cercanos a 1 representan una opini√≥n positiva mientras que el otro extremo una posici√≥n negativa.

```python
import matplotlib.pyplot as plt 

subjetividad = df['subjetividad']
polaridad = df['polaridad']

data = [subjetividad, polaridad]

plt.boxplot(data)
plt.title("Subjetividad/objetividad vs Polaridad")
plt.show()
```
