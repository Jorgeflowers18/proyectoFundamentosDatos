# Regresión Lineal


## 1. Regresión Lineal Simple.  Aplicado a un conjunto de datos generado automáticamente

- Source: https://www.iartificial.net/regresion-lineal-con-ejemplos-en-python/

```python
import numpy as np
import matplotlib.pyplot as plt 

%matplotlib inline 
```

### a. Generando el conjunto de datos (x,y)

```python

np.random.seed(42) # para poder reproducirlo

# Función para generar datos según una función lineal, siendo a = 1.25 y b = 0.1
def f(x):
    
    # f(x) = 1.25 + 0.1*x + 0.2*Ruido_Gaussiano
    y = 1.25 + 0.1*x + 0.2*np.random.randn(x.shape[0])
    return y

x = np.arange(0, 20, 0.5) # generamos valores x de 0 a 20 en intervalos de 0.5
x
```

```python
y = f(x) # calculamos y a partir de la función que hemos generado
# hacemos un gráfico de los datos que hemos generado
plt.scatter(x,y,label='data', color='blue')
plt.title('Datos');

print(y)
```

### b. Generando el modelo mediante el método de mínimos cuadrados

```python
# Importamos la clase de Regresión Lineal de scikit-learn
from sklearn.linear_model import LinearRegression #Regresión Lineal con scikit-learn
 
regresion_lineal = LinearRegression() # creamos una instancia de LinearRegression

# instruimos a la regresión lineal que aprenda de los datos (x,y)
regresion_lineal.fit(x.reshape(-1,1), y) 

# Parámetros que ha estimado la regresión lineal:
print('a = ' + str(regresion_lineal.intercept_) , 'y  b = ' + str(regresion_lineal.coef_) )

```

Existen diferencias en el valor de los coeficientes:

- Para a se estima 1,29 en lugar de 1,25
- Para b se estima 0.092 en lugar de 0.1

Este pequeño error es normal debido a la cantidad de ruido gaussiano que hemos introducido y al hecho de que hay muy pocos datos. 

```python
# ¿Qué hubiera pasado con datos perfectamente distribuidos sobre la recta?

# Línea perfecta, sin ruido Gaussiano:

y2 = 1.25 + 0.1*x

plt.scatter(x,y2,label='data', color='blue')
plt.title('Datos');

regresion_lineal2 = LinearRegression()

regresion_lineal2.fit(x.reshape(-1,1), y2) 

# Coeficientes casi perfectos:
print('a = ' + str(regresion_lineal2.intercept_) , 'y  b = ' + str(regresion_lineal2.coef_) )
```

### c. Utilizando el modelo de regresión para predecir nuevos valores

```python
# Utilizando el modelo para predecir:

x_new = np.array([5]); print('x_new', x_new)

y_new = regresion_lineal.predict(x_new.reshape(-1,1))

print(y_new)
```

La estimación de la regresión lineal del modelo que acabamos de entrenar para x = 5 es y = 1.7449.

### d.  Evaluando la calidad de la regresión lineal

Se evalua la calidad del modelo aprendido usando: datos de entrenamiento y/o nuevos datos denominado test set (no utilizados durante el entrenamiento).

La evaluación implica evaluar la capacidad de generalización del modelo (nuevos datos) y la calidad del modelo midiendo:

- El error cuadrático medio

- El coeficiente de determinación R^2: El coeficiente de determinación determina la calidad del modelo para replicar los resultados, y la proporción de variación de los resultados que puede explicarse por el modelo (Wikipedia). El rango de R2 está entre 0 y 1, siendo 1 el mejor valor esperado. 


```python
# Error Cuadrático Medio: conjunto de datos de entrenamiento (train set)

# importamos el cálculo del error cuadrático medio (MSE)
from sklearn.metrics import mean_squared_error

# Predecimos los valores y para los datos usados en el entrenamiento
predict_train = regresion_lineal.predict(x.reshape(-1,1))

# Calculamos el Error Cuadrático Medio (MSE = Mean Squared Error)
mse = mean_squared_error(y_true = y, y_pred = predict_train)

# La raíz cuadrada del MSE es el RMSE
rmse = np.sqrt(mse)
print('Error Cuadrático Medio (MSE) = ' + str(mse))
print('Raíz del Error Cuadrático Medio (RMSE) = ' + str(rmse))
```

```python
# Error Cuadrático Medio: conjunto de datos de prueba (test set)

# conjunto de prueba:
x_test = np.arange(5, 10, 0.5)  # generar números de 5 a 10 con intervalos de 0.5
y_test = f(x_test)  # generar valores reales para y

# Predecimos los valores y para los datos usados en el entrenamiento
predict_test = regresion_lineal.predict(x_test.reshape(-1,1))

# Calculamos el Error Cuadrático Medio (MSE = Mean Squared Error)
mse = mean_squared_error(y_true = y_test, y_pred = predict_test)

# La raíz cuadrada del MSE es el RMSE
rmse = np.sqrt(mse)
print('Error Cuadrático Medio (MSE) = ' + str(mse))
print('Raíz del Error Cuadrático Medio (RMSE) = ' + str(rmse))
```

```python
# Coeficiente de determinación R^2:
#Para medir el coeficiente de determinación R2 de la regresión lineal usaremos el método score.

# r^2 sobre el dataset de entrenamiento:
print('Coeficiente de Determinación R2 (train) = ' + str(regresion_lineal.score(x.reshape(-1,1), y)))

# r^2 sobre el dataset de test:
print('Coeficiente de Determinación R2 (test) = ' + str(regresion_lineal.score(x_test.reshape(-1,1), y_test)))
```

¿Por qué el coeficiente es más bajo con el conjunto de test? Es normal que el modelo se comporte mejor con los datos de entrenamiento, puesto que son los datos que se utilizaron para construir el modelo. Cuando la calidad del modelo es peor para conjuntos de datos nuevos (test), podría ser porque:

- Existió sobre-entrenamiento (overfitting)
- Se necesitan más datos durante el training y test. Aumentando el test set a un longitud mayor se consigue un R^2 de 0,611

### d.  Visualización de los resultados

Para saber cómo se está comportando el modelo. La línea que el modelo ha aprendido siguiendo el método de los mínimos cuadrados aparece en el siguiente gráfico en rojo.

```python
import pandas as pd

pd.Series(y_test.tolist()).plot(color='steelblue')  # Valores reales
pd.Series(predict_test.tolist()).plot(marker="o", linewidth=0, color='black') # Valores generados por el modelo
```

## 2. Regresión Lineal Múltiple.  Aplicado a un conjunto de datos de sklearn: The Boston Dataset

- Source: https://medium.com/qu4nt/el-machine-learning-y-la-regresión-python-c94db2968a4e


- Para ver la descripción del dataset ir a https://towardsdatascience.com/machine-learning-project-predicting-boston-house-prices-with-regression-b4e47493633d

The features can be summarized as follows:

- CRIM: This is the per capita crime rate by town
- ZN: This is the proportion of residential land zoned for lots larger than 25,000 sq.ft.
- INDUS: This is the proportion of non-retail business acres per town.
- CHAS: This is the Charles River dummy variable (this is equal to 1 if tract bounds river; 0 otherwise)
- NOX: This is the nitric oxides concentration (parts per 10 million)
- RM: This is the average number of rooms per dwelling
- AGE: This is the proportion of owner-occupied units built prior to 1940
- DIS: This is the weighted distances to five Boston employment centers
- RAD: This is the index of accessibility to radial highways
- TAX: This is the full-value property-tax rate per 10000
- PTRATIO: This is the pupil-teacher ratio by town
- B: This is calculated as 1000(Bk — 0.63)², where Bk is the proportion of people of African American descent - by town
- LSTAT: This is the percentage lower status of the population
- MEDV: This is the median value of owner-occupied homes in $1000s

### a. Carga del dataset: Boston houses prices

```python
from sklearn.datasets import load_boston 
boston = load_boston()
print(boston.data.shape)  # 506 observaciones y 13 variables

# variables independientes
x = boston.data #data

# variable dependiente o resultado:
y = boston.target 

# nombres de variables 
features = boston.feature_names ; features
```

### b. Verificar los supuestos para aplicar regresión múltiple

```python
# 1. ¿Existe una relación lineal entre la variable dependiente y cada una de las variables independientes?
import seaborn as sns

plt.scatter(x[:, 0], y, label='data', color='blue')  # CRIM: This is the per capita crime rate by town
```

```python
plt.scatter(x[:, 4], y, label='data', color='blue') # NOX: This is the nitric oxides concentration (parts per 10 million)
```

```python
data_DF = pd.DataFrame(x, )
data_DF = pd.merge(data_DF, pd.DataFrame(y), right_index=True, left_index=True)


data_DF.columns = ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD',
       'TAX', 'PTRATIO', 'B', 'LSTAT', 'PRICE']
data_DF
```

```python
# Coeficientes de correlación:

corrMatrixM = data_DF.corr()

corrMatrixM.PRICE.sort_values(axis=0, ascending=True)  # Cada variable independiente vs. PRICE (variable dependiente)

```

```python
# Relación líneal positiva más alta:

plt.scatter(x[:, 5], y, label='data', color='blue') # RM: Average number of rooms per dwelling
```

```python
# Relación líneal negativa :

plt.scatter(x[:, 12], y, label='data', color='blue') # LSTAT: Percentage lower status of the population
```

### c. Preparar los dos conjuntos de datos: entrenamiento y prueba:

```python

# Doc: https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.train_test_split.html
from sklearn.model_selection import train_test_split as tts

# 80% de las observaciones quedan para el conjunto de entrenamiento y 20% para el conjunto de prueba:

x_ent, x_test, y_ent, y_test = tts(x, y, train_size=0.8, random_state=42, shuffle=True)

print(x_ent.shape, x_test.shape)
```

```python
# Crear una instancia y entrenar el modelo con los conjuntos de datos de entrenamiento:

lm = LinearRegression()
lm.fit(x_ent, y_ent)
```

```python
# Coeficientes de la función de regresión:
print('a = ' + str(lm.intercept_) , 'y  B = ' + str(lm.coef_) )
```

```python
# Utilizar el modelo para generar las predicciones:

y_predict = lm.predict(x_test)

print(y_predict)
```

```python
x_test[0,:]
```

```python
# Verificar el R^2 (sobre el conjunto de test):

lm.score(x_test, y_test )  # con 1 aprendió muy bien.
```

```python
# Calculamos el Error Cuadrático Medio (MSE = Mean Squared Error)
mse = mean_squared_error(y_true = y_test, y_pred = y_predict)

# La raíz cuadrada del MSE es el RMSE
rmse = np.sqrt(mse)
print('Error Cuadrático Medio (MSE) = ' + str(mse))
print('Raíz del Error Cuadrático Medio (RMSE) = ' + str(rmse))
```

```python
# Visualización valores reales vs predichos por el modelo de regresión:

pd.Series(y_test.tolist()).plot(color='steelblue')
pd.Series(y_predict.tolist()).plot(marker="o", linewidth=0, color='black')
```

### d. Cómo mejorar el modelo: Feature selection


* Método: Variance inflation factor (VIF)

La identificación de la multicolinealidad se puede realizar mediante el método Variance inflation factor (VIF). 

VIF indica el porcentaje de la varianza inflada para el coeficiente de cada variable.

- Un valor hasta de 1 indica que no hay colinealidad
- Un VIF entre 1 y 5 indica una colinealidad moderada
- Mientras que los valores superiores a 5 indican una alta colinealidad. 

Algunos casos en los que sería aceptable un VIF alto incluyen el uso de términos de interacción, términos polinomiales o variables ficticias (variables nominales con tres o más categorías). Las matrices de correlación permiten la identificación de correlación entre pares de variables, mientras que VIF permite la evaluación general de la multicolinealidad.

El VIF se puede calcular usando el paquete statsmodels.


 Source and documentation:
 - https://towardsdatascience.com/collinearity-measures-6543d8597a2e
 - https://towardsdatascience.com/how-to-identify-the-right-independent-variables-for-machine-learning-supervised-algorithms-439986562d32
 - https://github.com/AngShengJun/petProj/blob/master/eda_viz/Viz_lib_LR_prototyped_1.ipynb

```python

from statsmodels.stats.outliers_influence import variance_inflation_factor

#  Calcular the VIF para todo el dataset (si ya se conoce cuál es la variable dependiente, se elimina esa variable)

vifF1 = pd.DataFrame()


vifF1["features"] = data_DF.columns
vifF1["vif_Factor"] = [variance_inflation_factor(data_DF.values, i) for i in range(data_DF.shape[1])]
print(vifF1)
```

En general, deberíamos apuntar a un VIF de menos de 10 para las variables independientes.

Cuando no sabemos qué variable es la independiente, la variable con el VIF más alto podría ser un candidato potencial para la variable independiente. En este caso se elimina la variable se calculan nuevamente los VIF.
En el ejemplo analizado, RM podría ser una potencial variable independiente.

En nuestro caso, ya sabemos cuál es la variable que queremos predecir así que la eliminamos a la variable y recalculamos el VIF.


```python

#  Calcular the VIF para cada variable independiente:

vifF2 = pd.DataFrame()

x_DF2 = data_DF.drop('PRICE', axis=1)
x_DF2.head()

vifF2["features"] = x_DF2.columns
vifF2["vif_Factor"] = [variance_inflation_factor(x_DF2.values, i) for i in range(x_DF2.shape[1])]
print(vifF2)
```

Al eliminar el PRICE, el VIF de algunos otros potenciales independientes también disminuyó: LSTAT y RM, variables con las cuáles tenía el valor más alto del coeficiente de correlación.

- Selección de variables independientes:  

Eliminar las variables con multicolinealidad e identificar la lista de variables independientes que son relevantes para predecir los precios de las casas de Boston.

Podríamos eliminar varias variables a la vez (con el valor VIF más alto) o podríamos ir probando una a una. Tenemos que realizar varias iteracione shasta llegar a tener VIFs con hasta un valor de 10.

```python
x_DF2 = x_DF2.drop('PTRATIO', axis=1)
x_DF2.head()
```

```python
vifF2 = pd.DataFrame()
vifF2["features"] = x_DF2.columns
vifF2["vif_Factor"] = [variance_inflation_factor(x_DF2.values, i) for i in range(x_DF2.shape[1])]
print(vifF2)
```

```python
x_DF2 = x_DF2.drop('NOX', axis=1)
vifF2 = pd.DataFrame()
vifF2["features"] = x_DF2.columns
vifF2["vif_Factor"] = [variance_inflation_factor(x_DF2.values, i) for i in range(x_DF2.shape[1])]
print(vifF2)
```

```python
x_DF2 = x_DF2.drop('TAX', axis=1)
vifF2 = pd.DataFrame()
vifF2["features"] = x_DF2.columns
vifF2["vif_Factor"] = [variance_inflation_factor(x_DF2.values, i) for i in range(x_DF2.shape[1])]
print(vifF2)
```

```python
x_DF2 = x_DF2.drop('RM', axis=1)
vifF2 = pd.DataFrame()
vifF2["features"] = x_DF2.columns
vifF2["vif_Factor"] = [variance_inflation_factor(x_DF2.values, i) for i in range(x_DF2.shape[1])]
print(vifF2)
```

```python
x_DF2 = x_DF2.drop('AGE', axis=1)
vifF2 = pd.DataFrame()
vifF2["features"] = x_DF2.columns
vifF2["vif_Factor"] = [variance_inflation_factor(x_DF2.values, i) for i in range(x_DF2.shape[1])]
print(vifF2)
```

```python
x_DF2 = x_DF2.drop('B', axis=1)
vifF2 = pd.DataFrame()
vifF2["features"] = x_DF2.columns
vifF2["vif_Factor"] = [variance_inflation_factor(x_DF2.values, i) for i in range(x_DF2.shape[1])]
print(vifF2)
```

Como podemos ver que después de varias iteraciones tenemos VIF de todas las variables restantes menos de 10. Hemos eliminado las variables con multicolinealidad y hemos identificado la lista de variables independientes que son relevantes para predecir los precios de las casas de Boston.

-  Crear un nuevo modelo con las variables relevantes:

```python
# 80% de las observaciones quedan para el conjunto de entrenamiento y 20% para el conjunto de prueba:

x_new = data_DF.drop(['PTRATIO', 'NOX', 'TAX', 'RM', 'AGE', 'B', 'PRICE'], axis=1)

x_ent, x_test, y_ent, y_test = tts(x_new, y, train_size=0.8, random_state=42, shuffle=True)

print(x_ent.shape, x_test.shape)


# Crear una instancia y entrenar el modelo con los conjuntos de datos de entrenamiento:

lmI = LinearRegression()
lmI.fit(x_ent, y_ent)


# Coeficientes de la función de regresión:
print('a = ' + str(lmI.intercept_) , 'y  B = ' + str(lmI.coef_) )

# Utilizar el modelo para generar las predicciones:

y_predict = lmI.predict(x_test)

# Verificar el R^2 (sobre el conjunto de test):

print(40*'*')
print('Coeficiente R^2: ', lmI.score(x_test, y_test)) 
print(40*'*')


# Calculamos el Error Cuadrático Medio (MSE = Mean Squared Error)
mse = mean_squared_error(y_true = y_test, y_pred = y_predict)

# La raíz cuadrada del MSE es el RMSE
rmse = np.sqrt(mse)
print('Error Cuadrático Medio (MSE) = ' + str(mse))
print('Raíz del Error Cuadrático Medio (RMSE) = ' + str(rmse))

# Visualización valores reales vs predichos por el modelo de regresión:

pd.Series(y_test.tolist()).plot(color='steelblue')
pd.Series(y_predict.tolist()).plot(marker="o", linewidth=0, color='black')
```

### e. Guardar el modelo de forma persistente

Cuando consigamos un resultado aceptable, podemos guardar el modelo para reutilizarlo directo para clasificar un nuevo conjunto de datos

```python
# Guardar el modelo de regresión original con las 13 variables independientes:

import joblib

joblib.dump(lm, "Boston_model_o.pkl")
```

```python
# Utilizar el modelo entrenado que hemos guardado

lm = joblib.load("Boston_model_o.pkl")
lm.score(boston.data, boston.target)  # R^2 = 0.73
```

```python
x_short = boston.data[:, [0, 1, 2, 3, 7, 8, 12]]  # seleccionar solo los features relevantes

lmI.score(x_short, boston.target)  # R^2 = 0.73
```
