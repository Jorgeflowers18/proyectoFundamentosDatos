```python
import pandas as pd
deda = pd.read_csv('../proyect/kaggle_dataset/world-happiness-report.csv')
```

```python
deda = deda.rename(columns={'Country name': 'country_name',\
                  'Log GDP per capita': 'gdp_per_capita',\
                  'Life Ladder':'life_Ladder',\
                   'Healthy life expectancy at birth':'healthy_life_expectancy_at_birth',\
                   'Social support':'social_supp', \
                   'Positive affect':'postve_affect', 'Negative affect':'negtve_affect',\
                   'Freedom to make life choices':'freedom_to_make_life_choices',\
                   'Perceptions of corruption': 'percept_of_corruption'
                  })
lista_paises = deda['country_name']
```

```python
l1 = []
j = 0
l_countries = deda['country_name']
for i in deda['country_name'].unique():
    j = j+1
    l1.append(j)

df_countries = pd.DataFrame({'id_countries': l1,'countries': deda['country_name'].unique()})
deda['country'] = l_countries
df_countries
```
