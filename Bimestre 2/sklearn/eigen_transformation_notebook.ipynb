```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pylab import rcParams
%matplotlib inline
rcParams['figure.figsize'] = 8,8
```

$$
\quad
\begin{pmatrix} 
9 & 4 \\
4 & 3 
\end{pmatrix}
$$

```python
x = np.linspace(-1,1,100)
```

```python
y1 = np.sqrt(1 - np.square(x))
y2 = -1 * y1
```

```python
plt.plot(x,y1, 'b')
plt.plot(x,y2, 'b')
plt.xlim([-2, 2])
plt.ylim([-2, 2])
plt.show()
```

```python
def transformation(x,y):
    return 9*x + 4*y, 4*x + 3*y
```

```python
x_new1, y_new1 = transformation(x,y1)
x_new2, y_new2 = transformation(x,y2)
```

```python
plt.plot(x_new1,y_new1, 'r')
plt.plot(x_new2,y_new2, 'r')
```

```python
eig_vals, eig_vecs = np.linalg.eig(np.array([[9,4],[4,3]]))
print('Eigenvectors \n%s' %eig_vecs)
print('\nEigenvalues \n%s' %eig_vals)
```

```python
soa = np.array([[0, 0, 
                 eig_vals[0] * eig_vecs[0][0], 
                 eig_vals[0] * eig_vecs[1][0]]])

soa1 = np.array([[0, 0, 
                  eig_vals[1] * eig_vecs[0][1], 
                  eig_vals[1] * eig_vecs[1][1]]])

X, Y, U, V = zip(*soa)
X1, Y1, U1, V1 = zip(*soa1)

plt.plot(x,y1, 'b')
plt.plot(x,y2, 'b')
plt.quiver(X, Y, U, V, angles='xy', scale_units='xy', scale=1)
plt.quiver(X1, Y1, U1, V1, angles='xy', scale_units='xy', scale=1)
plt.xlim([-1, 10])
plt.ylim([-1, 10])
plt.show()
```

```python
soa = np.array([[0, 0, 
                 eig_vals[0] * eig_vecs[0][0], 
                 eig_vals[0] * eig_vecs[1][0]]])

soa1 = np.array([[0, 0, 
                  eig_vals[1] * eig_vecs[0][1], 
                  eig_vals[1] * eig_vecs[1][1]]])

X, Y, U, V = zip(*soa)
X1, Y1, U1, V1 = zip(*soa1)
plt.plot(x_new1,y_new1, 'r')
plt.plot(x_new2,y_new2, 'r')
plt.quiver(X, Y, U, V, angles='xy', scale_units='xy', scale=1)
plt.quiver(X1, Y1, U1, V1, angles='xy', scale_units='xy', scale=1)
plt.draw()
plt.show()
```

```python
x = np.arange(0,10)
```

```python
print ("Eigen Vectors for Eigen values {} = {} and {}".format(eig_vals[0], 
                                                             eig_vecs[0][0], 
                                                             eig_vecs[1][0]))
```

```python
y = (eig_vecs[1][0] * x )/ eig_vecs[0][0]
```

```python
xnew, ynew = transformation(x,y)
soa = np.array([[0, 0, 
                 eig_vals[0] * eig_vecs[0][0], 
                 eig_vals[0] * eig_vecs[1][0]]])
X, Y, U, V = zip(*soa)

plt.quiver(X, Y, U, V, angles='xy', scale_units='xy', scale=1)
plt.scatter(xnew, ynew, color=['red'])
plt.draw()
plt.show()
```
